```python
import pandas as pd
```

```python
data_1=pd.read_csv("Potion_type_and_dialogues.csv", quoting=3)
```

```python
data_1['Potion'].unique()
```

```python
data_1['Dialogue']
```

```python
import re
import nltk
```

```python
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer
from nltk.stem import LancasterStemmer
nltk.download('wordnet')
```

```python
words=stopwords.words("english")
for word in words:
    if ("n't" or "not") in word:
        words.remove(word)
words.remove("not")
words.remove('no')
```

```python
print(words)
```

```python
cleaned_dialouges=[]
for i in range(0,559):
    dialouge=re.sub('[^a-zA-Z]', ' ', data_1['Dialogue'][i])
    dialouge=dialouge.lower()
    dialouge=dialouge.split()
    ps=PorterStemmer()
    # lt=WordNetLemmatizer()
    # ls=LancasterStemmer()
    dialouge=[ps.stem(word) for word in dialouge if not word in words]
    dialouge=' '.join(dialouge)
    cleaned_dialouges.append(dialouge)
```

```python
cleaned_dialouges
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cvr=CountVectorizer(max_features = 1500)
x=cvr.fit_transform(cleaned_dialouges).toarray()
```

```python
x
```

```python
y=data_1['Potion'].values
```

```python
y
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier()
```

```python
from sklearn.model_selection import GridSearchCV
parameters = [{'n_estimators':[10,20,30,40,50,60,100], 'criterion':['gini','entropy','log_loss'], 'max_features':['sqrt', 'log2', None]
               , 'class_weight':['balanced', 'balanced_subsample', None]}]
grid_search = GridSearchCV(estimator = classifier,
                           param_grid = parameters,
                           scoring = 'accuracy',
                           cv = 10,
                           n_jobs = -1)
grid_search.fit(x, y)
```

```python
grid_search.best_score_
```

```python
grid_search.best_params_
```
